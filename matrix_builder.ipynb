```python
df_matrix_A_aligned_copy = df_matrix_A_aligned
```

```python
# Max value of each row is saved in a new column max_value
df_matrix_A_aligned_copy['max_value'] = df_matrix_A_aligned_copy.max(axis=1)
```

```python
df_matrix_A_aligned_copy['max_value']
```

```python
# index and max_value column are separated into a new dataframe
d_matrix = df_matrix_A_aligned_copy[['max_value']].copy()
```

```python
# Replace d_matrix values as 0 -> 1 and 1 -> 0
d_matrix_1_replaced = d_matrix.replace(1, -1)
d_matrix_0_replaced = d_matrix_1_replaced.replace(0, 1)
d_matrix_replaced = d_matrix_0_replaced.replace(-1, 0)
d_matrix.shape
```

```python
d_matrix.head()
```

```python
# Converting d_matrix_replaced to numpy matrix
d_mat = d_matrix_replaced.values
```

```python
df_matrix_A_aligned.index
```

```python
# Add a new column as prob to get initial probability 1/n
df_matrix_A_aligned_copy['prob'] = 1/7349
```

```python
df_matrix_A_aligned_copy['prob']
```

```python
# Separating index and prob columns to a new dataframe
p_matrix = df_matrix_A_aligned_copy[['prob']].copy()
```

```python
p_matrix.head()
```

```python
# Building p0 matrix from p_matrix dataframe with equal initial probabilities
p_mat = p_matrix.values
```

```python
p_mat
```

```python
# Building e matrix as [1,1,1,......,1] with dimensions 1x7349 
e_matrix = np.ones(7349).reshape([1,7349])
```

```python
e_matrix
```

```python
e_matrix.shape
```

```python
p0 = p_mat
```

```python
p0.shape
```

```python
# Running the probability calculation iterations
for i in range(20):
    p1 = np.dot(p0.T, a_matrix)+np.dot(np.dot(p0.T, d_mat), e_matrix/7349)
    p1 = p1/np.linalg.norm(p1)
    p0 = p1.T
```

```python
p1
```

```python
p0
```

```python
# Get dataframe back from final probabilities
df = pd.DataFrame(data=p0,index=df_matrix_A_aligned.index,columns=p_matrix.columns)
```
